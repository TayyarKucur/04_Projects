Find out the types values of the columns and save the result as a dataframe.

```python
df_type = df.dtypes.to_frame(name="dtypes")
df_type
```

```python
df_notnull = df.notnull().sum().to_frame(name = "notnull")
df_notnull
```

```python
df_isnull = df.isnull().sum().to_frame(name="isnull")
df_isnull
```

Find out the unique counts of the columns and save the result as a dataframe

```python
df_nunique = df.nunique().to_frame(name="nunique")
df_nunique
```

```python
type_k = df_type.reset_index()
notnull_k = df_notnull.reset_index()
isnull_k = df_isnull.reset_index()
nunique_k = df_nunique.reset_index()
merge1 = pd.merge(type_k,notnull_k,how='inner', on='index')
merge2 = pd.merge(merge1,isnull_k,how='inner', on='index')
merge_end = pd.merge(merge2,nunique_k,how='inner', on='index')
merge_end.rename(columns={"index":"Columns"})
```
